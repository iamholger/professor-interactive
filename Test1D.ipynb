### A simple 1D parametrisation study, comparing a function with polynomials of different orders

#### The function

```python
import numpy as np
def f(x):
    return 3*x**3 + 2*x +1
```

#### Some data generation

```python
A = np.linspace(0, 10, 20)
Y = f(A)
```

#### The plot shows the 'true' function values as red dots. The different graphs show predictions from parametrisations of those points for different orders of polynomials 

```python
% matplotlib notebook
```

```python
import professor2 as prof
X = np.array([[a] for a in A])
ipols = [prof.Ipol(X, Y, i, "order%i" % i) for i in xrange(0,7)]
Z = [[i.value(x) for x in X] for i in ipols]

import pylab
pylab.plot(A, Y, "ro", label="True function values")
for num, z in enumerate(Z):
    pylab.plot(X, z, "-", label="Ipol order %i" % (num))
pylab.legend(loc=2)
pylab.show()
```
